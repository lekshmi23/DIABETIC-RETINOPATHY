```python
import pickle
import warnings
warnings.filterwarnings('ignore')
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import svm
from  sklearn.metrics import plot_confusion_matrix 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import svm



feat_path='/content/drive/MyDrive/NEW DR FOLDER/feature.pkl'
label_path='/content/drive/MyDrive/NEW DR FOLDER/label.pkl'
f = open(feat_path, 'rb')
feat=pickle.load(f)
f.close()
f = open(label_path, 'rb')
label=pickle.load(f)
f.close()

X=np.array(feat)
y=np.array(label)
X_train, X_test, y_train, y_test = train_test_split(
             X, y, test_size = 0.2)

#clf = svm.SVC()
clf = svm.LinearSVC()
clf.fit(X_train, y_train)


print(clf.score(X_train, y_train))
plot_confusion_matrix(clf,X_train, y_train)
plt.title('SVM Classifier - Confusion Matrix - Training')
plt.show()
print(clf.score(X_test, y_test))
plot_confusion_matrix(clf,X_test, y_test)
plt.title('SVM Classifier - Confusion Matrix - Testing')
plt.show()
```

```python
print('Total Feature Size',X.shape)
print('Abnormal' ,np.sum(y))
print('Normal' ,y.shape[0]-np.sum(y))
print('Train Size',X_train.shape)
print('Abnormal' ,np.sum(y_train))
print('Normal' ,y_train.shape[0]-np.sum(y_train))
print('Test Size',X_test.shape)
print('Abnormal' ,np.sum(y_test))
print('Normal' ,y_test.shape[0]-np.sum(y_test))
```

```python
from sklearn import metrics
confusion = metrics.confusion_matrix(y_test,clf.predict(X_test) )
print(confusion)

#print(metrics.accuracy_score(dib_train.Diabetes, dib_train.Diabetes_predicted))
```

```python
TP = confusion[1,1] # true positive 
TN = confusion[0,0] # true negatives
FP = confusion[0,1] # false positives
FN = confusion[1,0] # false negatives
```

```python
total1=sum(sum(confusion))
Accuracy = (confusion[0,0]+confusion[1,1])/total1
Specificity = confusion[0,0]/(confusion[0,0]+confusion[0,1])
Sensitivity = confusion[1,1]/(confusion[1,0]+confusion[1,1])
print('Accuracy :',Accuracy)
print(' Specificity:',Specificity)
print('Sensitivity :',Sensitivity)
```
